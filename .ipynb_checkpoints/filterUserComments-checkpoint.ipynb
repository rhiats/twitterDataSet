# Read User Comments

```python
users_arr = pickle.load( open( "english_speaking_authors.p", "rb" ) )
```

```python
user_content_df = pickle.load( open( "user_content.p", "rb" ) )
```

```python
def userComments(df,username):
    """
        Given a dataframe containing screennames and user_comments,
        return dataframe containing the userID, screenname, and comments for a specified user.
    """
    d = df[(df['screen_name'] == username)]
    
    return d[["user_id","screen_name","text"]]

def removeEmptyText(df):
    """
        Given a dataframe containing user_comments,
        return dataframe that does not have empty text cells.
    """
    d = df[(df['screen_name'] == username)]
    
    return d[["screen_name","text"]]
```

```python
for u in users_arr:
    d = userComments(user_content_df,u)
    if len(d) > 2: print(d)
```
